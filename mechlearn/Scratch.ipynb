```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def npmi(p1,p2,p12):
    pmi = np.log(p12/(p1*p2))
    return pmi/-np.log(p12)
```

```python
px = 0.1
py = 0.1
vs = []
pxy = np.linspace(px*py,min(px,py),50)
npmis = npmi(px,py,pxy)

plt.plot(pxy/min(px,py),npmis,'b')



px = 0.01
py = 0.01
pxy = np.linspace(px*py,min(px,py),50)
npmis = npmi(px,py,pxy)
plt.plot(pxy/max(px,py),npmis,'k')


px = 0.02
py = 0.05
pxy = np.linspace(px*py,min(px,py),50)
npmis = npmi(px,py,pxy)
plt.plot(pxy/max(px,py),npmis,'r')


px = 0.5
py = 0.5
pxy = np.linspace(px*py,min(px,py),50)
npmis = npmi(px,py,pxy)
plt.plot(pxy/min(px,py),npmis,'g')
plt.show()
```
